```python

import numpy as np


import matplotlib.pyplot as plt 


def mi_funcion(Vm, A,frec, fase, N, frecADC):
    
    ts = 1/frecADC
    tt = np.linspace(0,(N-1)*ts, N)
    ff = Vm + A*np.sin(2*np.pi*frec*tt + fase)
    return tt, ff
   
```

Graficamos los vectores que retorna la funcion 


```python
tt, ff = mi_funcion(Vm=50, A=5,frec=10, fase=np.pi/2, N=100, frecADC=500)
plt.figure()
grafico= plt.plot(tt, ff, 'ro')
plt.title('senoidal')
plt.xlabel('tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.show()
```

Podemos ver que al variar la frecuencia del ADC el grafico va cambiando, volviendose menos claro a medida que disminuye

```python
tt, ff = mi_funcion(Vm=50, A=5,frec=10, fase=np.pi/2, N=100, frecADC=50)
plt.figure()
grafico= plt.plot(tt, ff, 'ro')
plt.title('senoidal')
plt.xlabel('tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.show()
```

```python
tt, ff = mi_funcion(Vm=50, A=5,frec=10, fase=np.pi/2, N=100, frecADC=800)
plt.figure()
grafico= plt.plot(tt, ff, 'ro')
plt.title('senoidal')
plt.xlabel('tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.show()
```

Asi mismo, al variar la cantidad de muestras obtenemos mas o menos ciclos de la senoidal

```python
tt, ff = mi_funcion(Vm=50, A=5,frec=10, fase=np.pi/2, N=400, frecADC=500)
plt.figure()
grafico= plt.plot(tt, ff, 'ro')
plt.title('senoidal')
plt.xlabel('tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.show()
```
